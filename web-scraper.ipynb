```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def extract_source(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:50.0) Gecko/20100101 Firefox/50.0'}
    source=requests.get(url, headers=headers)
    return source
```

```python
URL = "https://www.rugsusa.com/rugsusa/control/search-rugs?size_grp=6x9"
page = extract_source(URL)
soup = BeautifulSoup(page, 'html.parser')
```

```python
first_page = soup.find(id='ui_target')
#print(first_page.prettify())
last_page_elem = first_page.find('span',class_="suffix")
last_page = last_page_elem.text
last_page = [int(s) for s in last_page.split() if s.isdigit()][0]
```

```python
if last_page > 10: 
    max_page = 10 
else:
    max_page = last_page
```

```python
df_scraped_prices = pd.DataFrame({'rug_name':[],
                                  'msrp_price':[],
                                  'listed_price':[],
                                  
                                 })
```

```python
data = []

for i in range(1,max_page+1):
    URL = "https://www.rugsusa.com/rugsusa/control/search-rugs?size_grp=6x9&p={page_num}".format(page_num = i)
    page = extract_source(URL)
    soup = BeautifulSoup(page, 'html.parser')
    results = soup.find(id='searchTarget')
    rug_listings = results.find_all('div',class_="resultTile prod productCard productinfoimagewithbadgesmask")
    for rug_listing in rug_listings:
        name_elem = rug_listing.find('div',class_='productname1 head-text')
        msrp_price_elem = rug_listing.find('span', class_='msrp body-text')
        default_price_elem = rug_listing.find('span', class_='default_price accent-text')
        num_review_elem = rug_listing.find('div', class_ = 'amountofreviews')
        print(name_elem.text)
        print("msrp price: ", msrp_price_elem.text)
        print(num_review_elem.text)
        print("listed price: ", default_price_elem.text, end='\n'*2)
        data.append([name_elem.text, msrp_price_elem.text, default_price_elem.text,num_review_elem.text])
```

```python
df = pd.DataFrame(data, columns=["rug_name","msrp_price","list_price","num_reviews"])
df
```
